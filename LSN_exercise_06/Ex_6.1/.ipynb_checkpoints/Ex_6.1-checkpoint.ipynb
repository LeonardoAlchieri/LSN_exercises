I run the programm multiple times: I want to record energy, heat capacity and susceptibity for different temperatures ```0.5 < T < 3```. This first part is done for ```h=0```.

````python
points=50
start =0.
T = np.linspace(0.2,3.0,num=points)
for i in range(len(T)):
    if( T[i] > 0.5):
        print(i)
        start = i
        break
print(T[start])
````

````python
y_ene = np.zeros([len(T),2])
y_heat = np.zeros([len(T),2])
y_mag = np.zeros([len(T),2])
y_chi = np.zeros([len(T),2])

err_ene = np.zeros([len(T),2])
err_heat = np.zeros([len(T),2])
err_mag = np.zeros([len(T),2])
err_chi = np.zeros([len(T),2])
````

````python
for element in range(2):
    if(element == 0):
        print("GIBBS")
    elif(element == 1):
        print("\n METROPOLIS")
    for i in range(start, len(T)):
        print('--- Progress', int((i-start+1)/(len(T)-start) * 100), '%', end='\r')
        temp = T[i]
        #print('Temperature used:', temp)
        sb.call('./clean.sh', shell=True)
        #print('./Monte_Carlo_ISING_1D.exe %f %f' %(temp, element))
        os.system('./Monte_Carlo_ISING_1D.exe %f %f' %(temp, element))
        
        #print('Programm run')
        f_ene = np.loadtxt("output.ene.0")

        f_heat = np.loadtxt("output.heat.0")
        
        f_mag = np.loadtxt("output.mag.0")

        f_chi = np.loadtxt("output.chi.0")

        y_ene[i][element] = f_ene[-1,2]
        y_heat[i][element] = f_heat[-1,2]
        y_mag[i][element] = f_mag[-1,2]
        y_chi[i][element] = f_chi[-1,2]
        
        err_ene[i][element] = f_ene[-1,3]
        err_heat[i][element] = f_heat[-1,3]
        err_mag[i][element] = f_mag[-1,3]
        err_chi[i][element] = f_chi[-1,3]
````

````python
figure(figsize=(10,7), dpi=100)

plt.errorbar(T[start:], y_ene[start:,0], yerr=err_ene[start:,0], color='blue', label='Gibbs')
plt.errorbar(T[start:], y_ene[start:,1], yerr=err_ene[start:,1], color='red', label='Metropolis')
plt.title('Internal Energy')
plt.legend(loc='best')
plt.xlabel('T')
plt.ylabel('U')
plt.show()
````

````python
figure(figsize=(10,7), dpi=100)

plt.errorbar(T[start:], y_heat[start:,0], yerr=err_heat[start:,0], color='blue', label='Gibbs')
plt.errorbar(T[start:], y_heat[start:,1], yerr=err_heat[start:,1], color='red', label='Metropolis')
plt.title("Heat Capacity")
plt.legend(loc='best')
plt.xlabel('T')
plt.ylabel('C')
plt.show()
````

````python
figure(figsize=(10,7), dpi=100)

plt.errorbar(T[start:], y_chi[start:,0], yerr=err_chi[start:,0], color='blue', label="Gibbs")
plt.errorbar(T[start:], y_chi[start:,1], yerr=err_chi[start:,1], color='red', label='Metropolis')
plt.title('Susceptibility')
plt.legend(loc='best')
plt.xlabel("T")
plt.ylabel('χ')
plt.show()
````

In order to evaluate some form of magnetisation, the same process as above is done for ```h=0.2```. 

````python
system('./change_h_0.02.sh')
````

````python
for element in range(2):
    if(element == 0):
        print("GIBBS")
    elif(element == 1):
        print("\n METROPOLIS")
    for i in range(0, len(T)):
        print('--- Progress', int((i+1)/(len(T)) * 100), '%', end='\r')
        temp = T[i]
        
        sb.call('./clean.sh', shell=True)
        
        os.system('./Monte_Carlo_ISING_1D.exe %f %f' %(temp, element))
                
        f_mag = np.loadtxt("output.mag.0")
        y_mag[i][element] = f_mag[-1,2]
        err_mag[i][element] = f_mag[-1,3]
````

````python
figure(figsize=(10,7), dpi=100)

plt.errorbar(T, y_mag[0:,0], yerr=err_mag[0:,0], color='blue', label='Gibbs')
plt.errorbar(T, y_mag[0:,1], yerr=err_mag[0:,1], color='red', label='Metropolis')
plt.title('Magnetisation for h=0.02')
plt.legend(loc='best')
plt.xlabel('T')
plt.ylabel('M')
plt.show()
````
