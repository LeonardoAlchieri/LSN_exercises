# Ex 2.1

- Compute the following 1D integral via Monte Carlo 
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$


1. sampling a uniform distribution in $[0,1]$
2. using importance sampling (i.e. sampling a non-uniform probability in $[0,1]$)

<span style="color:red">Show a picture of both your estimations of $I$</span> and their uncertainties with a large number of *throws* $M$ (e.g. $M\ge 10^4$) as a function of the number of blocks, $N$</font>.

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure 
from time import time
```

```python
start = time()
```

```python
f = np.loadtxt("part_1.txt", skiprows=1)
f2 = np.loadtxt("part_2.txt", skiprows=1)
```

I print the evaluation of the integral using a Uniform distribution and _importance sampling_ with $2(1-x)$ as distribution.

```python
figure(figsize=(10,7), dpi=200)

plt.errorbar(f[:,0],f[:,1],yerr=f[:,2], color='blue', label='Uniform')
plt.errorbar(f[:,0],f2[:,0],yerr=f2[:,1], color='green', label='$2(1-x)$')
plt.axhline(1, color='red', label='Real Value')
plt.legend(loc="upper right")
plt.xlabel("# throw")
plt.ylabel('I')
plt.grid(True)
plt.show()
```

As expected, the calculations with the _importance sampling_ tend to be far more precise then a Uniform distribution.

```python
end = time()

print("Total computing time:", int((end-start)*100)/100.,"sec.")
```
