# Ex 3.1

By using the following parameters: 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

compute at time $t=0$ **via Monte Carlo** the European call-option price, $C[S(0),0]$, and put-option price, $P[S(0),0]$:

1. <span style="color:red"> By sampling **directly** the final asset price $S(T)$</span> for a $GBM(r,\sigma^2)$
2. <span style="color:red"> By sampling the discretized $GBM(r,\sigma^2)$ path of the asset price</span> 
dividing $[0,T]$ in $10^2$ time intervals: $S(0)$, $S(t_1)$, $S(t_2)$, ... up to $S(t_{100}=T=1)$

<span style="color:red">Make four pictures for the estimation of the European call-option prices, $C[S(0),0]$ (direct and discretized), and put-option prices, $P[S(0),0]$ (direct and discretized), with their uncertainties </span> with a large number of asset prices at time $t=T$, say $M\ge 10^4$, as a function of the number of blocks, $N$</font>. <span style="color:blue"> As usual, in computing statistical uncertainties, use data blocking.</span>

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
from time import time
from os import system
```

```python
start = time()
```

```python
# execute the c++ programm
system("./main.exe")
```

```python
f1 = np.loadtxt('part1.txt', skiprows=2)
f2 = np.loadtxt('part2.txt', skiprows=2)
```

```python
figure(figsize=(7,5), dpi=100)

plt.errorbar(f1[:,0],f1[:,1], yerr=f1[:,2], color='blue')
plt.axhline(14.975790778311286, color='red', label='Black-Scholes')
plt.title('Direct Call')
plt.xlabel("Block")
plt.ylabel("$<call>$")
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
figure(figsize=(7,5), dpi=100)

plt.errorbar(f1[:,0],f1[:,3], yerr=f1[:,4], color='green')
#plt.errorbar(x,f[:,1],yerr=0)
plt.axhline(5.4595325819072364, color='red', label='Black-Scholes')
plt.title('Direct Put')
plt.xlabel("Block")
plt.ylabel("$<put>$")
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
figure(figsize=(7,5), dpi=100)

plt.errorbar(f2[:,0],f2[:,1], yerr=f2[:,2], color='blue')
#plt.errorbar(x,f[:,1],yerr=0)
plt.axhline(14.975790778311286, color='red', label='Black-Scholes')
plt.title('Descrete Call')
plt.xlabel("Block")
plt.ylabel("$<call>$")
plt.legend(loc='best')
plt.grid(True)

plt.show()
```

```python
figure(figsize=(7,5), dpi=100)

plt.errorbar(f2[:,0],f2[:,3], yerr=f2[:,4], color='green')
#plt.errorbar(x,f[:,1],yerr=0)
plt.axhline(5.4595325819072364, color='red', label='Black-Scholes')
plt.title('Descrete Put')
plt.xlabel("Block")
plt.ylabel("$<put>$")
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
end = time()

print("Total computing time:", int((end-start)*100.)/100., 'sec.')
```
