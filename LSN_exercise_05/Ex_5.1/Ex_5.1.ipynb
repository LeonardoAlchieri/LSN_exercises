```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.pyplot import figure
from time import time
```

```python
start = time()
```

```python
figure(figsize=(7,5), dpi=100)

f1 = np.loadtxt('results/1s_distances.txt', skiprows=1)
x1 = np.arange(len(f1))

plt.errorbar(x1, f1[:,0], yerr=f1[:,1])
plt.title('Average Radius for 1s in H')
plt.xlabel('blocks')
plt.ylabel('<r>')
plt.axhline(1.5, color='red', label='Analitical Value')
plt.legend(loc='best')
plt.show()
```

```python
f = np.loadtxt("results/3Dmap_1s.txt")
X=f[:,0]
Y=f[:,1]
Z=f[:,2]
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
plt.title('1s orbital in H atom')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(30, 30)
plt.savefig('plots/3dmap_1s.png', dpi=1200)
plt.show()
```

```python
f2 = np.loadtxt('results/2p_distances.txt', skiprows=1)
x2 = np.arange(len(f2))

figure(figsize=(7,5), dpi=100)

plt.errorbar(x2, f2[:,0], yerr=f2[:,1])
plt.title('Average Radius for 2p in H')
plt.xlabel('blocks')
plt.ylabel('<r>')
plt.axhline(5, color='red', label='Analytical Value')
plt.legend(loc='best')
plt.show()
```

```python
f1 = np.loadtxt("results/3Dmap_2p.txt")
X1=f1[:,0]
Y1=f1[:,1]
Z1=f1[:,2]
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X1, Y1, Z1, c=Z1, marker='.')
plt.title('2p orbital in H atom')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(15, 45)
plt.savefig('plots/3dmap_2p.png', dpi=600)
plt.show()
```
