<span style="color:red">Include the calculation of the radial distribution function, $g(r)$, as a function of the radial distance $r$ among particles, with $r \in \left[ 0; L/2\right]$ inside the Monte Carlo NVT code</span> we have provided: average value of $g(r)$ in each block (with output in the file: output.gofr.0), final average value of $g(r)$ with statistical uncertainties (with output in the file: output.gave.0).

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import time
start = time.time()
```

```python
n_blocks = 10
n_steps = 1000
```

## Solid

```python
state = 'solid'
```

```python
os.system('./clean.sh')
print('./Monte_Carlo_NVT.exe input.'+str(state)+' '+str(n_blocks) + ' ' + str(n_steps))
os.system('./Monte_Carlo_NVT.exe input.'+str(state)+' '+str(n_blocks) + ' ' + str(n_steps))
```

```python
ave_solid = np.loadtxt('output.gave.0')
```

## Liquid

```python
state = 'liquid'
```

```python
os.system('./clean.sh')
print('./Monte_Carlo_NVT.exe input.'+str(state)+' '+str(n_blocks) + ' ' + str(n_steps))
os.system('./Monte_Carlo_NVT.exe input.'+str(state)+' '+str(n_blocks) + ' ' + str(n_steps))
```

```python
ave_liquid = np.loadtxt('output.gave.0')
```

## Gas

```python
state = 'gas'
```

```python
os.system('./clean.sh')
print('./Monte_Carlo_NVT.exe input.'+str(state)+' '+str(n_blocks) + ' ' + str(n_steps))
os.system('./Monte_Carlo_NVT.exe input.'+str(state)+' '+str(n_blocks) + ' ' + str(n_steps))
```

```python
ave_gas = np.loadtxt('output.gave.0')
```

--------

```python
plt.style.use('classic')
plt.figure(num=None, figsize=(10, 8), dpi=600, facecolor='w', edgecolor='k')

plt.subplot(311)
plt.errorbar(ave_solid[:,0], ave_solid[:,1], yerr=ave_solid[:,2])
plt.xlabel('r/')
plt.ylabel('g(r)')
plt.title('Solid')

plt.subplot(312)
plt.errorbar(ave_liquid[:,0], ave_liquid[:,1], yerr=ave_liquid[:,2])
plt.xlabel('r/')
plt.ylabel('g(r)')
plt.title('Liquid')

plt.subplot(313)
plt.errorbar(ave_gas[:,0], ave_gas[:,1], yerr=ave_gas[:,2])
plt.xlabel('r/')
plt.ylabel('g(r)')
plt.title('Gas')

plt.subplots_adjust(hspace=0.8)

plt.show()
```

```python
end = time.time()
print('-- Time for computation: ', int(end-start), "sec")
```
