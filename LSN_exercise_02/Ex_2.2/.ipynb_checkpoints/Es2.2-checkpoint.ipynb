# Ex 2.2

1. on a cubic lattice with lattice constant $a=1$; at each discrete time the walker makes a forward or backward step of length equal to $a$ in one of the 3 principal directions of the lattice: $x$, $y$ or $z$
2. in the continuum; at each discrete time the walker makes a step of length equal to $a(=1)$ along a **random direction** obtained by sampling **uniformly** the solid angle: $\theta \in [0,\pi]$ and $\phi \in [0,2\pi]$

<span style="color:red">Show a picture of $\langle |\vec{r}_N| \rangle_{RW}$ for both RWs</span>, with their statistical uncertainties, as function of the step $i\in [0,10^2]$.

Note that you could try to fit both results with a function like $f(N)=k\sqrt{N}$. Do your results indicate a diffusive behavior?

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
from time import time
```

```python
start = time()
```

```python
f = np.loadtxt('dati.txt', skiprows=1)
```

```python
figure(figsize=(7,5), dpi=100)

plt.errorbar(f[:,0],f[:,1], yerr=f[:,2], color='#6f7b9f')
plt.title('Descrete')
plt.xlabel("step")
plt.ylabel("$<r>$")
plt.grid(True)
plt.show()
```

```python
figure(figsize=(7,5), dpi=100)

plt.errorbar(f[:,0],f[:,3], yerr=f[:,4], color='#6fa49f')
plt.title('Continuous')
plt.xlabel("step")
plt.ylabel("$<r>$")
plt.grid(True)
plt.show()
```

As expected, the results follow a rule $f(N) \sim \sqrt{N}$, where $N$ is the step. Both the approaches give the same outcome, as predicted.

```python
end = time()

print("Total computing time:", int((end-start)*100)/100.,"sec.")
```
