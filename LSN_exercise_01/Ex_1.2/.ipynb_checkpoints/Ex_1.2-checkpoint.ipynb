# Ex 1.2

1. <font color="red">Add two probability distributions</font> by using the **method of the inversion of the cumulative distribution** to sample from a **generic** <font color="red">exponential distribution</font>, $p(x) = \lambda \exp(-\lambda x)$, $x\in [0;+\infty]$ (see <a href="https://en.wikipedia.org/wiki/Exponential_distribution">this Wikipedia link</a>), and a **generic** <font color="red">Cauchy-Lorentz distribution</font> $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$, $x\in [-\infty;+\infty]$ (see <a href="https://en.wikipedia.org/wiki/Cauchy_distribution">this Wikipedia link</a>).
2. <font color="red">Make 3 pictures</font> with the histograms obtained filling them with $10^4$ realizations of $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ (for $N=1, 2, 10, 100$), being $x_i$ a random variable sampled throwing a *standard* dice (fig.1), an *exponential* dice (fig.2, use $\lambda=1$) and a *Lorentzian* dice (fig.3, use $\mu=0$ and $\Gamma=1$).

Note tha you can try to fit the case $N=100$ with a Gaussian for standard and exponential dices, whereas you should use a Cauchy-Lorentz distribution for the last case.

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import math
from matplotlib.pyplot import figure
from time import time
```

```python
start = time()
```

```python
x = np.loadtxt("data.txt")
```

```python
figure(figsize=(10,6), dpi=200)

width = 0.03
ALPHA = 0.5
plt.hist(x[:,1], bins=np.arange(min(x[:,1]), max(x[:,1])+width,width), alpha=ALPHA, label="N=1")
plt.hist(x[:,4], bins=np.arange(min(x[:,4]), max(x[:,4])+width,width), alpha=ALPHA, label="N=2")
plt.hist(x[:,7], bins=np.arange(min(x[:,7]), max(x[:,7])+width,width), alpha=ALPHA, label="N=10")
plt.hist(x[:,10], bins=np.arange(min(x[:,10]), max(x[:,10])+width,width), alpha=ALPHA, label="N=100")
plt.title("Exponential distribution")
plt.legend(loc='upper right')
plt.xlabel("x")
plt.ylabel("P(x)")
plt.show()
```

```python
figure(figsize=(10,6), dpi=200)

width_l = 0.1
plt.hist(x[:,2], bins=np.arange(min(x[:,2]), max(x[:,2])+width_l,width_l), alpha=0.5, label="N=1")
plt.hist(x[:,5], bins=np.arange(min(x[:,5]), max(x[:,5])+width_l,width_l), alpha=0.5, label="N=2")
plt.hist(x[:,8], bins=np.arange(min(x[:,8]), max(x[:,8])+width_l,width_l), alpha=0.5, label="N=10")
plt.hist(x[:,11], bins=np.arange(min(x[:,11]), max(x[:,11])+width_l,width_l), alpha=0.5, label="N=100")
#plt.axis([-5,5,0,1100])
plt.title("Cauchy-Lorentz distribution")
plt.legend(loc='upper right')
plt.xlabel("x")
plt.ylabel("P(x)")
plt.show()
```

```python
figure(figsize=(10,6), dpi=200)

width_uni = 0.02
plt.hist(x[:,0], bins=np.arange(min(x[:,0]), max(x[:,0])+width_uni,width_uni), alpha=0.5, label="N=1")
plt.hist(x[:,3], bins=np.arange(min(x[:,3]), max(x[:,3])+width_uni,width_uni), alpha=0.5, label="N=2")
plt.hist(x[:,6], bins=np.arange(min(x[:,6]), max(x[:,6])+width_uni,width_uni), alpha=0.5, label="N=10")
plt.hist(x[:,9], bins=np.arange(min(x[:,9]), max(x[:,9])+width_uni,width_uni), alpha=0.5, label="N=100")
#plt.axis([0,1,0,1500])
plt.title("Uniform distribution")
plt.legend(loc='upper right')
plt.xlabel("x")
plt.ylabel("P(x)")
plt.show()
```

_Nota: gli intervalli non coincidono tra di loro (all'interno degli stessi istogrammi). Si  preferito procedere in questa maniera per permettere una migliore visualizzazione dei dati_
